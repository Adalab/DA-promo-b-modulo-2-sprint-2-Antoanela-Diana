
Objetivos:

Usad la librería sidetable para:
1. Explorar los valores nulos.


```python
df.stb.missing()
```

2. Explorad la columna de country y reportad los resultados solo para aquellas categorías que incluyan el 70% de los datos.


```python
df.stb.freq(['country'], thresh= 70)

#En este Df vemos que el 70% de nuestros datos con respeto a los ataques de tiburones corresponden a Usa y Australia
```

3. Explorad la columna country en relación al sexo, filtrando de nuevo por aquellas categorías que incluyan el 70% de los datos.


```python
df.stb.freq(['country', 'genero_bien'], thresh= 70)
#>Con respeto al genero el 70% de nuestros datos reflejan que el genero masculino es el que mas sufre ataques en los paises Usa, Australia y South Africa.
```

4. Haced los mismo para la columna activity en relación a la columna fatal. En este caso filtrad y mostrad los datos que incluyan el 60% de los datos.


```python
df.stb.freq(['activity', 'fatal_bien'], thresh= 60)

#En general vemos que la mayoria de los ataques no son mortales.
#Las tres principales actividades donde mas ataques no fatales hay son surfing, swimming y spearfishing. 
#Esta tabla nos ayuda ver que las actividades con mayor numero de ataques es surfing con un 24.5% siendo estos no fatales, mientras que la actividad con mayor ataques fatales es nadar, pero solo con un 5% fatales
```
