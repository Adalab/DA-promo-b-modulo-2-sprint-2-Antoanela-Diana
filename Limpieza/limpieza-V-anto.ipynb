```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sidetable as stb
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer
```

```python
pd.options.display.max_rows=None
```

```python
df=pd.read_csv('../Datos/attacks_limpieza_III.csv')
df.head(2)
```

```python
df1= df.copy()
```

Es el momento de eliminar los nulos:
1. Reemplazad los valores nulos de la columna age por la media de la edad usando el m√©todo SimpleImputer.


```python
imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
imputer = imputer.fit(df1[['edades']])
df1['edades']=imputer.transform(df1[['edades']])
```

```python
df1.edades.unique()
```

```python
df1.edades.isnull().sum()
```

Reemplazad los valores nulos de la columna sex por la moda, usando el m√©todo SimpleImputer.üí° Pista üí° La moda en este tipo de aproximaci√≥n se indica como most_frequent.



```python
imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
imputer = imputer.fit(df1[['genero_bien']])
df1['genero_bien']=imputer.transform(df1[['genero_bien']])
```

```python
df1.genero_bien.isnull().sum()
```

Reemplazad los valores nulos de la columna type por el valor m√°s frecuente (la moda) con el m√©todo SimpleImputer.


```python
imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
imputer = imputer.fit(df1[['type']])
df1['type']=imputer.transform(df1[['type']])
```

```python
df1.type.isnull().sum()
```

Utilizad el m√©todo KNN Imputer para reemplazar todos los valores nulos de las columnas num√©ricas.


```python
numericas = df1.select_dtypes(include = np.number)
numericas
```

```python
imputerKNN = KNNImputer(n_neighbors=5)
imputerKNN.fit(numericas)
numericas_knn= imputerKNN.transform(numericas)
```

```python
df_knn_imputer = pd.DataFrame(numericas_knn, columns = numericas.columns)
```

```python
df_knn_imputer.head()
```

```python
columnas_knn = df_knn_imputer.columns
columnas_knn
```

Utilizad el m√©todo Iterative Imputer para reemplazar todos los valores nulos de las columnas num√©ricas.


```python
imputer = IterativeImputer(n_nearest_features=None, imputation_order='ascending')
imputer.fit(numericas)
imputer.transform(numericas)
numericas_trans = pd.DataFrame(imputer.transform(numericas), columns = numericas.columns)
numericas_trans.head()
```

```python
numericas_trans.isnull().sum()
```

```python
columnas = numericas_trans.columns
df.drop(columnas, axis = 1, inplace = True)
df[columnas] = numericas_trans[columnas]
df.head(2)
```

```python
nulos = pd.DataFrame((df.isnull().sum() * 100) / df.shape[0]).reset_index()
nulos.columns = ["columna", "porcentaje"]
nulos
```

¬øPodr√≠ais explicar qu√© diferencia hay entre estos dos √∫ltimos m√©todos?


>IterativeImputer, que modela cada caracter√≠stica con valores perdidos como una funci√≥n de otras variables, y utiliza esa estimaci√≥n para la imputaci√≥n. Lo hace de forma iterada: en cada paso, una columna 
se designa como salida(como variable que queremos predecir)  y las otras columnas se tratan como entradas X (como predictoras, es decir, las que nos van a ayudar a predecir 
los valores nulos de la columna que queremos). Se ajusta un regresor en (X, y). Se devuelven los resultados de la √∫ltima ronda de imputaci√≥n.
El IterativeImputer utiliza los datos disponibles en otras columnas con el fin de estimar los valores nulos que se imputan o queremos quitar.

>KNN IMPUTER solo lo podremos usar para variables numericas o de formato fecha.
